### histogram 

```python
# 3 histogram for order_hour_of_day
hist = df['order_hour_of_day'].plot.hist(bins = 50)
```

This histogram shows that most of the orders are ordered between the hours of 9 and 17 with hour 10 and 11 being the most

### Bar Chart 

```python
# 4 Bar chart for loyalty_flag
bar = df['loyalty_flag'].value_counts().plot.bar()
```

This bar graph shows that there are more regular customers shopping 

### Line Chart 

```python
#creating random sample of dataframe with 70/30
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
#splitting dataframe into 2 samples bid storing 70% and small storing 30%
big = df[dev]
small = df[~dev]
```

```python
len(df)
```

```python
len(big) + len(small)
```

```python
#dataframe only containing columns needed for visual
df_line = small[['prices','order_hour_of_day']]
```

```python
#create line graph for prices and order_hour_of_day
line = sns.lineplot(data = df_line, x = 'order_hour_of_day',y = 'prices')
```

This line graph shows that the times that customers spend the most are in the hours of 4 and 8

### Scatterplot

```python
#7 scatterplot of age and income
scatterplot = sns.scatterplot(x = 'Age', y = 'income',data = df)
```

This scatterplot shows that regardless of age most customers' income is $200,000 or less. The scatterplot also shows that customers less then 40 do not make over $400,000. There is a significant number of customers 40-80 that make $300,000 and less.
